First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# Let's create a bar chart for the sucess rate of each orbit. # HINT use groupby method on Orbit column and get the mean of Class column

orbit_success_rate = df.groupby('Orbit')['Class'].mean()

plt.figure(figsize=(12, 6))
orbit_success_rate.plot(kind='bar')
plt.xlabel('Orbit Type', fontsize=20)
plt.ylabel('Success Rate', fontsize=20)
plt.title('Success Rate by Orbit Type', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.

### TASK  5: Visualize the relationship between Payload and Orbit type

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date
def Extract_year(date):
    return date.apply(lambda x: x.split("-")[0])
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate

df['Year'] = Extract_year(df['Date'])

# Calculate the average success rate for each year
yearly_success_rate = df.groupby('Year')['Class'].mean()

# Plot the line chart
plt.figure(figsize=(12, 6))
yearly_success_rate.plot(kind='line')
plt.xlabel('Year', fontsize=20)
plt.ylabel('Average Success Rate', fontsize=20)
plt.title('Launch Success Yearly Trend', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

You can observe that the success rate since 2013 kept increasing till 2017 (stable in 2014) and after 2015 it started increasing.

```python
# prompt: Find the dates of the first successful landing outcome on ground pad

# Filter the dataframe for successful landings on ground pad
successful_ground_landings = df[(df['Outcome'] == 'Success (ground pad)') & (df['Class'] == 1)]

# Get the dates of these successful landings
dates_of_successful_ground_landings = successful_ground_landings['Date']

# Print the dates
print("Dates of the first successful landing outcome on ground pad:")
for date in dates_of_successful_ground_landings:
    print(date)
```

```python
# prompt: Calculate the total number of successful and failure mission outcomes

# Calculate the value counts of the 'Class' column
outcome_counts = df['Class'].value_counts()

# Map the class labels to 'Success' and 'Failure'
# Assuming 'Class' = 1 is Success and 'Class' = 0 is Failure
outcome_labels = {1: 'Success', 0: 'Failure'}
outcome_counts.index = outcome_counts.index.map(outcome_labels)

# Print the results
print("Mission Outcome Counts:")
outcome_counts
```

```python
# prompt: # prompt: List the names of boosters which have successfully landed on drone ship and had payload mass greater than 4000 but less than 6000

# Filter for successful landings on drone ship
successful_drone_landings = df[(df['Outcome'] == 'Success (drone ship)') & (df['Class'] == 1)]

# Filter for payload mass greater than 4000 and less than 6000
filtered_boosters = successful_drone_landings[(successful_drone_landings['PayloadMass'] > 4000) & (successful_drone_landings['PayloadMass'] < 6000)]

# Get the names of the boosters
booster_names = filtered_boosters['BoosterVersion'].tolist()

# Print the booster names
print("Boosters that successfully landed on drone ship with payload mass between 4000 and 6000 kg:")
for name in booster_names:
name
```

```python
# prompt: List the names of the booster which have carried the maximum payload mass

# Find the maximum payload mass
max_payload_mass = df['PayloadMass'].max()

# Filter the dataframe to get rows with the maximum payload mass
boosters_with_max_payload = df[df['PayloadMass'] == max_payload_mass]

# Get the names of the boosters
booster_names = boosters_with_max_payload['BoosterVersion'].tolist()

# Print the names of the boosters
print("Boosters which have carried the maximum payload mass:")
for name in booster_names:
    print(name)
```

```python
# prompt: List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015

# Filter for failed landings on drone ship in 2015
failed_drone_landings_2015 = df[(df['Outcome'] == 'Failure (drone ship)') & (df['Year'] == '2015')]

# Select the required columns
result = failed_drone_landings_2015[['Outcome', 'BoosterVersion', 'LaunchSite']]

# Print the result
print("Failed landing outcomes on drone ship in 2015:")
result
```

```python
# prompt: Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Filter the dataframe by the date range
filtered_df = df[(df['Date'] >= '2010-06-04') & (df['Date'] <= '2017-03-20')]

# Count the occurrences of each landing outcome
outcome_counts_ranked = filtered_df['Outcome'].value_counts().sort_values(ascending=False)

# Print the ranked counts
print("Ranked count of landing outcomes between 2010-06-04 and 2017-03-20:")
outcome_counts_ranked
```
